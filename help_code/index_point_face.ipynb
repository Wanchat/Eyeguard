```python
import numpy as np
import cv2
import dlib
import matplotlib.pyplot as plt
from imutils import face_utils
```

```python
img = cv2.imread('face_woman.jpg')
# plt.imshow(img,cvtColor.COLOR_BGR2RGB)
# cv2.line(img, (0,100),(500,100),(0,255,0),2)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(r'C:\Users\Wanch\Google Drive\Thesis\code\help_code\shape_predictor_68_face_landmarks.dat') 
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rects = detector(gray, 0)

for rect in  rects:

    shape = predictor(gray, rect)
    shape = face_utils.shape_to_np(shape)

            # (x, y, w, h) = face_utils.rect_to_bb(rect)
            # cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 3)


    for (x,y) in shape:
        cv2.circle(img,( x, y), 1, (0, 0, 255), -1)
```
